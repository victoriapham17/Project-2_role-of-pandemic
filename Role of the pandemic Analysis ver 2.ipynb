```python
from scipy.stats import pearsonr

def calculate_pvalues(arizona_data):
    arizona_data = arizona_data.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=arizona_data.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in arizona_data.columns:
        for c in arizona_data.columns:
            pvalues[r][c] = round(pearsonr(arizona_data[r], arizona_data[c])[1], 4)
    return pvalues


arizona_data = data.corr()
arizona_data = arizona_data.round(2)
pval = calculate_pvalues(arizona_data)
# create three masks
r1 = arizona_data.applymap(lambda x: '{}*'.format(x))
r2 = arizona_data.applymap(lambda x: '{}**'.format(x))
r3 = arizona_data.applymap(lambda x: '{}***'.format(x))
# apply them where appropriate
arizona_data = arizona_data.mask(pval<=0.05,r1)
arizona_data = arizona_data.mask(pval<=0.01,r2)
arizona_data = arizona_data.mask(pval<=0.001,r3)
arizona_data
```
